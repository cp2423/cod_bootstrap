```python
import cv2
import numpy as np
from numpy import ndarray
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [16, 16]

TEST_FILES = [
    "/Users/chris/Dev/cod_records/aws/extracted/31829_B016711/31829_B016711-00000.jpg",
    "/Users/chris/Dev/cod_records/george circumstance of death.jpg"
]
```

```python
# crop off the left
fp = TEST_FILES[0]
img = cv2.imread(fp)
red = img[:,:,2]
thresh = cv2.threshold(red, 200, 255, cv2.THRESH_BINARY_INV)[1]

left_quarter = thresh.shape[1] // 4
print(left_quarter)
```

```python
fp = TEST_FILES[1]
img = cv2.imread(fp, cv2.IMREAD_GRAYSCALE)[30:80, 200:400]
img = cv2.blur(img, (3,3))
canny_output = cv2.Canny(img, 150, 255)
plt.imshow(canny_output)
contours, hierarchy = cv2.findContours(canny_output, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
```
