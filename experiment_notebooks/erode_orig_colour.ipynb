```python
import cv2
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [16, 16]

TEST_FILES = [
    "/Users/chris/Dev/cod_records/aws/extracted/31829_B016711/31829_B016711-00000.jpg",
    "/Users/chris/Dev/cod_records/george circumstance of death.jpg"
]
```

```python
# optional mapping of values with morphological shapes
def morph_shape(val):
    if val == 0:
        return cv2.MORPH_RECT
    elif val == 1:
        return cv2.MORPH_CROSS
    elif val == 2:
        return cv2.MORPH_ELLIPSE

def get_element(shape, size):
    ms = morph_shape(shape)
    return cv2.getStructuringElement(ms, (2 * size + 1, 2 * size + 1), (size, size))
```

```python
src = cv2.imread(TEST_FILES[0], cv2.IMREAD_GRAYSCALE)

fig, axs = plt.subplots(2)
i = 0

sizes = range(1, 3)
shape = 0

for size in sizes:
    element = get_element(shape, size)
    erosion_dst = cv2.erode(src, element, iterations=2)
    axs[i].imshow(erosion_dst, cmap='gray')
    i += 1
```
