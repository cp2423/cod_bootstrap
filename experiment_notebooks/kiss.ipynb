```python
import cv2
import numpy as np
from numpy import ndarray
from matplotlib import pyplot as plt
import pytesseract

plt.rcParams['figure.figsize'] = [16, 16]

TEST_FILES = [
    "/Users/chris/Dev/cod_records/aws/extracted/31829_B016711/31829_B016711-00000.jpg",
    "/Users/chris/Dev/cod_records/george circumstance of death.jpg"
]
```

```python
def denoise(img: ndarray) -> ndarray:
    # using recommended values from https://docs.opencv.org/3.4/d5/d69/tutorial_py_non_local_means.html
    return cv2.fastNlMeansDenoising(img, 10, 10, 7, 21)
```

```python
fig, axs = plt.subplots(2)
i = 0

for f in TEST_FILES:
    img = cv2.imread(f)
    red = img[30:60,:,2]
    thresh = cv2.threshold(red, 200, 255, cv2.THRESH_BINARY_INV)[1]
    clean = denoise(thresh)
    print(pytesseract.image_to_string(clean, config="--psm 7"))
    i += 1
```
