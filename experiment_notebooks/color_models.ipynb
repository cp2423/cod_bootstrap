```python
import cv2
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 16]
```

```python
TEST_FILES = [
    "/Users/chris/Dev/cod_records/aws/extracted/31829_B016711/31829_B016711-00000.jpg",
    "/Users/chris/Dev/cod_records/george circumstance of death.jpg"
]
```

```python
thresholds = [30, 60, 90, 120, 150, 180, 210]
fig, axs = plt.subplots(7, 2)

COMPONENT = 2

for i, threshold in enumerate(thresholds):
    for j, f in enumerate(TEST_FILES):
        src = cv2.imread(f)
        comp = cv2.cvtColor(src, cv2.COLOR_BGR2LUV)[:,:,COMPONENT]
        thresh = cv2.threshold(comp, threshold, 255, cv2.THRESH_BINARY_INV)[1]
        axs[i,j].imshow(thresh, cmap='gray')
        axs[i,j].set_title(f"Theshold={threshold} (component {COMPONENT})")
```
